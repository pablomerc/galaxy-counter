```python
import h5py
import numpy as np

path = '/data/vision/billf/scratch/pablomer/legacysurvey_hsc/preprocessed_hsc_legacy_metadata_8192.h5'
path_neighbors = '/data/vision/billf/scratch/pablomer/data/neighbours_v2.h5'

# 1. Load the search keys (hsc_ids)
with h5py.File(path, 'r') as f:
    hsc_ids = f['hsc_object_id'][:]
    # Optimization: Convert to set if hsc_ids is large, though np.isin handles this well
    # hsc_ids_set = set(hsc_ids) 

# 2. Load the target database
with h5py.File(path_neighbors, 'r') as f:
    neighbor_ids = f['object_id_hsc'][:]
    source_types = f['source_type'][:]

# 3. Create the mask on the ORIGINAL arrays to preserve indices
# Condition A: Is the neighbor ID in our search list?
mask_is_in_list = np.isin(neighbor_ids, hsc_ids)

# Condition B: Is the source type 0?
mask_is_valid_type = (source_types == 0)

# Combine conditions (Boolean AND)
final_mask = mask_is_in_list & mask_is_valid_type

# 4. Get the absolute indices in the file
# np.where returns the indices where the mask is True
indices_in_file = np.where(final_mask)[0]
matching_ids = neighbor_ids[indices_in_file]

# 5. Print results
# Now 'idx' corresponds to the actual row number in 'path_neighbors'
for fid, idx in zip(matching_ids, indices_in_file):
    print(f"ID: {fid}, Index in file: {idx}")
```
